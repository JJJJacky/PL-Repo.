```python
import csv
from IPython.display import display, HTML

# 读取CSV文件
filename = '南部老街.csv'
rows = []
with open(filename, 'r', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        rows.append(row)

# 创建可点击的超链接并对齐表格
html_table = "<table><tr>{}</tr>".format(
    ''.join(["<th>{}</th>".format(key) for key in rows[0].keys()]))

for row in rows:
    html_table += "<tr>"
    for col, val in row.items():
        if col == "Google Map":
            html_table += f'<td><a href="{val}" target="_blank">{val}</a></td>'
        else:
            html_table += f'<td>{val}</td>'
    html_table += "</tr>"

html_table += "</table>"

# 在Jupyter Notebook上显示表格
display(HTML(html_table))
```

```python
 import folium

# 创建地图对象
map_center = [23.560695764548296, 120.60303476599734]  # 地图中心点的经纬度坐标
zoom_level = 14  # 地图缩放级别
map_obj = folium.Map(location=map_center, zoom_start=zoom_level)

# 在地图上标记地址
address = "嘉義縣梅山鄉太平村"
location = [23.560695764548296, 120.60303476599734]  # 地址的经纬度坐标
name = "太平老街"
marker = folium.Marker(location=location, popup=address, tooltip=name)
marker.add_to(map_obj)

# 显示地图
map_obj
```
