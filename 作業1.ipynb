```python
#Q1
#語言使用英文的電影有幾部?
import pandas as pd

# 讀取CSV檔案，並將'original_language'欄位轉成字串型態
data = pd.read_csv('movies_metadata.csv', dtype={'original_language': str})

# 使用篩選器找出 'original_language' 欄位值為 'en' 的記錄
en_movies = data[data['original_language'] == 'en']

# 計算 'original_language' 欄位值為 'en' 的記錄數量
num_en_movies = len(en_movies)

# 印出結果（中文）
print(f"原始語言為 'en' 的電影數量為 {num_en_movies} 部。\n")
```

```python
#Q2
#電影語言使用非英文的有幾部?
import pandas as pd

# 讀取CSV檔案，並將'original_language'欄位轉成字串型態
data = pd.read_csv('movies_metadata.csv', dtype={'original_language': str})

# 使用篩選器找出 'original_language' 欄位值不為 'en' 的記錄
non_en_movies = data[data['original_language'] != 'en']

# 計算 'original_language' 欄位值不為 'en' 的記錄數量
num_non_en_movies = len(non_en_movies)

# 印出結果（中文）
print(f"原始語言不是 'en' 的電影數量為 {num_non_en_movies} 部。")
```

```python
#Q4
#在西元2000年後上映的電影有幾部?
import pandas as pd

# 讀取CSV檔案，並將 'release_date' 欄位轉成日期型態
data = pd.read_csv('movies_metadata.csv', parse_dates=['release_date'], dtype={'release_date': str})

# 使用篩選器找出 'release_date' 欄位在 2000 年 1 月 1 日之後的記錄
after_2000_movies = data[data['release_date'] >= '2000-01-01']

# 計算 'release_date' 欄位在 2000 年 1 月 1 日之後的記錄數量
num_after_2000_movies = len(after_2000_movies)

# 印出結果（中文）
print(f"在西元 2000 年之後上映的電影數量為 {num_after_2000_movies} 部。")
```

```python
#Q5
#在西元2000年前上映的電影有幾部?
import pandas as pd

# 讀取CSV檔案，並將 'release_date' 欄位轉成日期型態
data = pd.read_csv('movies_metadata.csv', parse_dates=['release_date'], dtype={'release_date': str})

# 使用篩選器找出 'release_date' 欄位在 2000 年 1 月 1 日之前的記錄
before_2000_movies = data[data['release_date'] < '2000-01-01']

# 計算 'release_date' 欄位在 2000 年 1 月 1 日之前的記錄數量
num_before_2000_movies = len(before_2000_movies)

# 印出結果（中文）
print(f"在西元 2000 年之前上映的電影數量為 {num_before_2000_movies} 部。")

```

```python
#Q6
#電影為T開頭的有哪些?
import pandas as pd

# 讀取CSV檔案
data = pd.read_csv('movies_metadata.csv')

# 使用篩選器找出 original_title 欄位為 T 開頭的所有記錄
t_movies = data[data['original_title'].str.startswith('T')]

# 列出所有符合條件的記錄
for title in t_movies['original_title']:
    print(title)
```
