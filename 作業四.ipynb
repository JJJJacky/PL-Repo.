```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import jieba

# 要爬取的目標網址
url = 'https://www.ptt.cc/bbs/WomenTalk/M.1637307390.A.907.html'

# 建立一個Session物件，用於保持連線狀態
session = requests.Session()

# 設定cookies，繞過PTT的年齡驗證
session.cookies.set('over18', '1')

# 發送GET請求取得網頁內容
response = session.get(url)

# 解析HTML
soup = BeautifulSoup(response.text, 'html.parser')

# 找到所有的評論元素
comments = soup.find_all('div', class_='push')

# 建立空的list來儲存評論內容
comment_list = []

# 逐一處理評論內容
for comment in comments:
    # 取得評論內容
    content = comment.find('span', class_='f3 push-content').text.strip()
    comment_list.append(content)

# 使用jieba進行斷詞
seg_list = []
for comment in comment_list:
    seg_list.append(jieba.lcut(comment))

# 將斷詞結果轉換為DataFrame
df = pd.DataFrame({'Comment': comment_list, 'Segmented': seg_list})

# 顯示DataFrame
print(df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import jieba

# 要爬取的目標網址
url = 'https://www.ptt.cc/bbs/WomenTalk/M.1637307390.A.907.html'

# 建立一個Session物件，用於保持連線狀態
session = requests.Session()

# 設定cookies，繞過PTT的年齡驗證
session.cookies.set('over18', '1')

# 發送GET請求取得網頁內容
response = session.get(url)

# 解析HTML
soup = BeautifulSoup(response.text, 'html.parser')

# 找到所有的評論元素
comments = soup.find_all('div', class_='push')

# 建立空的list來儲存評論內容和斷詞結果
data = []

# 逐一處理評論內容
for comment in comments:
    # 取得評論內容
    content = comment.find('span', class_='f3 push-content').text.strip()
    segmented = jieba.lcut(content)
    data.append([content, segmented])

# 將data轉換為DataFrame
df = pd.DataFrame(data, columns=['Comment', 'Segmented'])

# 顯示DataFrame
print(df)
```

```python
# 顯示斷詞結果
for i, row in df.iterrows():
    for j, word in enumerate(row['Segmented']):
        print(f"Segment {j+1}: {word}")
    print()
```

```python
from collections import Counter

# 計算詞語出現頻率
word_freq = Counter()
for row in df['Segmented']:
    word_freq.update(row)

# 按照詞語頻率進行排序
sorted_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# 儲存排序後的詞語及其頻率到新的list
sorted_list = []
for word, freq in sorted_freq:
    sorted_list.append((word, freq))

# 顯示排序後的詞語及其頻率
for word, freq in sorted_list:
    print(f"{word}: {freq}")
```
