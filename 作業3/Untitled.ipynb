```python
import requests
import csv
from bs4 import BeautifulSoup

url = 'https://news.ltn.com.tw/search?keyword=%E8%BB%8A%E7%A6%8D'
response = requests.get(url)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    titles = soup.find_all('a', class_='tit')
    
    with open('hw3.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow(['Title', 'Link']) # 寫入表頭
        
        for title in titles:
            title_text = title.text.strip()
            title_link = title.get('href')
            writer.writerow([title_text, title_link]) # 寫入資料

            
else:
    print('Failed to retrieve the webpage')
```

```python
import csv
import json

csv_file = 'hw3.csv'
json_file = 'hw3.json'

data = []
with open(csv_file, newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        data.append(row)

with open(json_file, 'w', encoding='utf-8') as jsonfile:
    json.dump(data, jsonfile, ensure_ascii=False, indent=4)
```

```python
import json

csvfile = open('hw3.csv', 'r', encoding='utf-8')
jsonfile = open('hw3.json', 'w', encoding='utf-8')

fieldnames = ("title", "link")
reader = csv.DictReader(csvfile, fieldnames)
out = json.dumps([row for row in reader])

jsonfile.write(out)

csvfile.close()
jsonfile.close()
```
