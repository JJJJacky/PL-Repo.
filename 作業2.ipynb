```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案
df = pd.read_csv('movies_metadata.csv')

# 取得vote_average欄位資料
vote_average_data = df['vote_average']

# 篩選出不為NaN的值
vote_average_data = vote_average_data.dropna()

# 設定直方圖區間數量
num_bins = 10

# 繪製直方圖
plt.hist(vote_average_data, num_bins)

# 設定標題和軸標籤
plt.title('Vote Average Distribution')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')

# 顯示圖表
plt.show()
```

```python
#Q1 該欄位vote_average資料中，出現頻率最高之數值區間為何?

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# 讀取資料
movies_metadata = pd.read_csv('movies_metadata.csv')

# 取得 vote_average 欄位的資料
vote_average_data = movies_metadata['vote_average']

# 設定直方圖的區間數量
bins = 10

# 建立直方圖
fig, ax = plt.subplots()
n, bins, patches = ax.hist(vote_average_data, bins=bins)

# 取得最大頻率的索引值
max_count_index = np.argmax(n)

# 取得答案區間的邊界值
bin_left, bin_right = bins[max_count_index], bins[max_count_index + 1]

# 輸出結果
print('最常出現 vote_average 的區間為 [{:.1f}, {:.1f})'.format(bin_left, bin_right))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案
df = pd.read_csv('movies_metadata.csv')

# 過濾掉original_language欄位中的NaN值
languages = df['original_language'].dropna()

# 計算各個語言出現的頻率
language_counts = languages.value_counts()

# 取得前10個出現頻率最高的語言
top_10_languages = language_counts[:10]

# 繪製直方圖
plt.bar(top_10_languages.index, top_10_languages.values)
plt.xlabel('Language')
plt.ylabel('Frequency')
plt.title('Top 10 Languages in movies_metadata')
plt.show()
```

```python
#Q2 欄位original_language中，使用ja語言的電影有幾部
# 取得ja這個語言出現的頻率
ja_frequency = language_counts['ja']

# 輸出結果
print("使用日本語言的電影有", ja_frequency, "部")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取CSV檔案
df = pd.read_csv('movies_metadata.csv')

# 將popularity欄位轉換為浮點數型別
df['popularity'] = pd.to_numeric(df['popularity'], errors='coerce')

# 選擇需要的欄位，過濾掉NaN值，根據popularity欄位遞減排序，並取前10筆資料
top_10_popularity = df[['original_title', 'popularity']].dropna().sort_values(by='popularity', ascending=False)[:10]

# 設定圖表大小
plt.figure(figsize=(10, 6))

# 繪製直方圖
plt.bar(top_10_popularity['original_title'], top_10_popularity['popularity'])
plt.xlabel('Movie Title')
plt.ylabel('Popularity')
plt.title('Top 10 Popular Movies in movies_metadata')
plt.xticks(rotation=90)
plt.show()

```

```python
#Q3 最受歡迎的電影是甚麼? 列出名稱和數值
# 取得第一受歡迎的電影
most_popular_movie = top_10_popularity.iloc[0]

# 列印電影名稱和y軸數值
print('Most Popular Movie:', most_popular_movie['original_title'])
print('Popularity:', most_popular_movie['popularity'])
```
