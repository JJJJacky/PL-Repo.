```python
import csv
import requests
from bs4 import BeautifulSoup

# 目標網頁的URL
url = 'https://www.ptt.cc/bbs/iOS/M.1664272513.A.B6B.html'

# 使用requests套件發送GET請求取得網頁資料
response = requests.get(url)

# 檢查是否成功取得網頁資料
if response.status_code == 200:
    # 使用BeautifulSoup解析HTML內容
    soup = BeautifulSoup(response.text, 'html.parser')

    # 找到文章內容的標籤
    article_content = soup.find(id='main-content')

    if article_content:
        # 取得文章標題
        title = article_content.find('span', class_='article-meta-value').text

        # 取得文章內容
        content = article_content.find_all('div', class_='article-metaline')[-1].find_next_siblings('div')
        content_text = ' '.join([c.text for c in content])

        # 輸出結果
        print('文章標題:', title)
        print('文章內容:')
        print(content_text)

        # 將結果寫入CSV檔案
        with open('result.csv', 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.writer(csvfile)
            writer.writerow(['文章標題', '文章內容'])
            writer.writerow([title, content_text])

        print('結果已成功寫入 result.csv 檔案')
    else:
        print('找不到文章')
else:
    print('無法取得網頁資料')
```

```python
import csv
import re

# 建立空的列表
comments = []

# 讀取CSV檔案並轉換為列表
with open('result.csv', 'r', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # 跳過標題欄位
    for row in reader:
        comment_with_username = row[1]  # 假設留言內容在第二個欄位
        comment = re.findall('\((.*?)\)', comment_with_username)[0]
        comments.append(comment)

# 顯示列表中的留言
for comment in comments:
    print(comment)

```

```python
import jieba

# 假設comments是包含留言的列表

# 建立空的斷詞結果列表
seg_list = []

# 使用jieba對每個留言進行斷詞
for comment in comments:
    seg_list.append(list(jieba.cut(comment)))

# 顯示斷詞結果
for segs in seg_list:
    print(segs)
```

```python
import jieba
from collections import Counter

# 假設 seg_list 是包含斷詞結果的列表

# 合併所有斷詞結果
all_words = [word for segs in seg_list for word in segs]

# 計算詞頻
word_counts = Counter(all_words)

# 根據詞頻進行排序
sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

# 顯示排序結果
for word, count in sorted_words:
    print(word, count)
```
