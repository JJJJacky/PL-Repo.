```python
#Q1
#標題adult的值為true的資料有幾筆?
# 讀取 CSV 檔案
df = pd.read_csv('movies_metadata.csv')
    
# 選擇 adult 欄位為 True 的列數
adult_true = df[df['adult'] == 'True']

# 計算筆數
count = len(adult_true)

# 輸出結果
print(f"計算標題 adult 的內容為 true 的資料有 {count} 筆")
```

```python
#Q2
#語言使用除了英文的電影共有幾部
# 讀取 CSV 檔案
df = pd.read_csv('movies_metadata.csv')

# 選擇 original_language 欄位不等於 en 的列數
not_en = df[df['original_language'] != 'en']

# 計算筆數
count = len(not_en)

# 輸出結果
print(f"original_language 欄位中除了 'en' 以外的資料有 {count} 筆")
```

```python
#Q3
#在西元2000年後發布的電影有幾部
# 讀取 CSV 檔案
df = pd.read_csv('movies_metadata.csv')

# 將 release_date 欄位轉換為日期格式
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')

# 選擇 release_date 欄位在 2000 年之後的列數
after_2000 = df[df['release_date'] >= pd.Timestamp(2000, 1, 1)]

# 計算筆數
count = len(after_2000)

# 輸出結果
print(f"release_date 欄位中在西元 2000 年以後的資料有 {count} 筆")
```
