```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ptt.cc/bbs/Gossiping/M.1675049783.A.171.html'

cookies = {
    'over18': '1'
}

response = requests.get(url, cookies=cookies)

if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # 提取帖子标题
    title_element = soup.find_all(class_='article-meta-value')
    if len(title_element) >= 3:
        title = title_element[2].text
    else:
        title = "未找到标题"
    
    # 提取帖子内容
    content_element = soup.find(id='main-content')
    content = content_element.text
    
    # 存储结果到列表
    result = [title, content]
    
    # 打印结果
    print('结果:', result)
else:
    print('请求失败')
```

```python
import jieba
import pandas as pd
from collections import Counter

# 假设result是您之前爬取并存储在列表中的结果
result = [title, content]

# 分词处理
seg_list = []
for text in result:
    seg_text = jieba.lcut(text)
    seg_list.extend(seg_text)

# 计算词语频率
word_freq = Counter(seg_list)

# 按频率进行排序
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# 转换为DataFrame
df = pd.DataFrame(sorted_word_freq, columns=['词语', '频率'])

# 打印结果
print(df)
```

```python
import jieba
import pandas as pd
from collections import Counter

# 读取停用词列表
stopwords_path = 'stopwords.txt'  # 停用词文件路径
stopwords = set()
with open(stopwords_path, 'r', encoding='utf-8') as f:
    for line in f:
        stopwords.add(line.strip())

# 假设result是您之前爬取并存储在列表中的结果
result = [title, content]

# 分词处理并过滤停用词
seg_list = []
for text in result:
    seg_text = jieba.lcut(text)
    seg_text = [word for word in seg_text if word not in stopwords]
    seg_list.extend(seg_text)

# 计算词语频率
word_freq = Counter(seg_list)

# 按频率进行排序
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# 转换为DataFrame
df = pd.DataFrame(sorted_word_freq, columns=['词语', '频率'])

# 显示前100项
df_top100 = df.head(100)
print(df_top100)
```
